*User-defined Functions*

```python
def berechnegesamtgehalt(grundgehalt=10000, bonus=0.01):
    gesamtgehalt = grundgehalt + grundgehalt * bonus
    return gesamtgehalt

print(berechnegesamtgehalt(grundgehalt=10, bonus=1))
```

```python
berechnegesamtgehalt = lambda x, y: x + x * y
berechnegesamtgehalt(1000, 0.1)
```

```python
#Anstatt
salary_list = [31268, 12936, 6132, 12936, 213609, 213970]
def calculatetotalsalary(base_salary=10000, bonus = 0.01):
  return base_salary * (1+bonus)
totalsalarylist = [calculatetotalsalary(salary) for salary in salary_list]

totalsalarylist
```

```python
#Lieber so
salary_list = [31268, 12936, 6132, 12936, 213609, 213970]
totalsalarylist = [(lambda x, y: x * (1+y))(salary, 0.01) for salary in salary_list]

totalsalarylist
```
