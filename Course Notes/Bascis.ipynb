**Funktionen**

```python
## Definition Function (the manipulator of data): a reusable piece of code that performs a specific task


# Aufbau einer function

def greet():
  return "Ficker"
greet()

job_title = "Data Analyst"
job_location = "Erfurt"
job_salary = 100000

def display_jobinformation(title, location, salary):
  return print(title, location, salary)
display_jobinformation(job_title, job_location, job_salary)
```

**Klassen**

```python

## Classes: Templates for creating objects

class JobPost:
  def __init__(self, title, location, salary):
    self.title = title
    self.location = location
    self.salary = salary

  def display_jobinformation(self):
    return print(self.title, self.location, self.salary)


JobPost(job_title, job_location, job_salary)
# with this piece of code we have created an instance, which we can now f.e. modify by passing attributes

job_3 = JobPost(job_title, job_location, job_salary)
job_3.title
```

**Methoden**

```python
## Methods: The functions of an object - a function defined inside a class that operates on its objects

# z.B. die __init__ method
job_3.display_jobinformation()
```
