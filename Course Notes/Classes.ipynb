```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate = 0.1, currency = "â‚¬"):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.currency = currency
    self.total_salary = base_salary * (1 + bonus_rate)
    self.bonus_rate = self.total_salary / base_salary - 1

  def __repr__(self):
    return f'{self.base_salary:,.0f}{self.currency}'

  def calculate_total_salary(self):
    return f'{self.total_salary:,.0f}{self.currency}'

  def calculate_bonus_rate(self):
    return f'{round(self.bonus_rate,2)}{"%"}'

```

```python
salary = BaseSalary(1000)
salary.calculate_bonus_rate()
```
