```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
lutschi = df.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc='median')
lutschi.reset_index(inplace=True)
lutschi.set_index('salary_year_avg', inplace=True)
lutschi.head()
```
