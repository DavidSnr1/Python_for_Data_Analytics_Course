```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_salary_by_title_and_country = df[df['job_title_short'] == 'Data Scientist']
df_salary_by_title_and_country= df_salary_by_title_and_country.pivot_table(values='salary_year_avg', index=['company_name', 'job_country'], aggfunc='median')
filtered_pivot_table = df_salary_by_title_and_country[df_salary_by_title_and_country['salary_year_avg'] > 200000]
filtered_pivot_table.head(10)
```
