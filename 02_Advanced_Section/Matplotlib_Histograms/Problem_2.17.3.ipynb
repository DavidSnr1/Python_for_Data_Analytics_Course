```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
from matplotlib import cm
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_de = df[df['job_title_short'] == 'Data Engineer']
df_de['salary_year_avg'].plot(kind='hist', bins=50, color='red')

mean_salary = df_de['salary_year_avg'].mean()
median_salary = df_de['salary_year_avg'].median()

plt.axvline(mean_salary, color='green', linestyle= 'dashed', label=  f'Mean: ${mean_salary:,.0f}')
plt.axvline(median_salary, color='blue', linestyle= 'dashed', label=  f'Mean: ${mean_salary:,.0f}')
```
