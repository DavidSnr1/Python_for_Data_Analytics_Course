```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_filtered = df.dropna(subset=['salary_year_avg']).copy()
```

```python
def set_salary_category(row):
    if row["salary_year_avg"] < 60000:
        return "Low"
    elif row["salary_year_avg"] < 100000:
        return "Medium"
    else:
        return "High"
    
    
df_filtered["salary_category"] = df_filtered.apply(set_salary_category, axis=1)
```
