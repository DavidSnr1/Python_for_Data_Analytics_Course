```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_oct = df[df['job_posted_date'].dt.month_name() == 'October']
df_nov = df[df['job_posted_date'].dt.month_name() == 'November']
df_dec = df[df['job_posted_date'].dt.month_name() == 'December']
```

```python
df_4_quarter = pd.concat([df_oct, df_nov, df_dec])
df_4_quarter
    
```
