```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_with_salary = df.dropna(subset='salary_year_avg')
print("Length of dataframe          :", len(df))
print("Length of cleaned dataframe  :", len(df_with_salary))
print("Makes a drifference of       :", len(df)-len(df_with_salary))
```
