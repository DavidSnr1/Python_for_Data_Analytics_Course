```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_company_salary = df.groupby('company_name')['salary_year_avg'].mean().reset_index()
df_company_remotes = df[df['job_work_from_home'] == True]
df_company_remotes
df_company_remotes= df.groupby('company_name')['job_work_from_home'].count().reset_index()
df_company_remotes.rename(columns={'job_work_from_home' : 'remote_count'}, inplace=True)
df_merged = pd.merge(left=df_company_salary, right=df_company_remotes, on='company_name', how='inner')
df_merged.head()
```
