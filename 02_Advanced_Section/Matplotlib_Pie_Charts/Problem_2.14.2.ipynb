```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
from matplotlib import cm
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_pie = df[(df['job_schedule_type'] == 'Full-time') | (df['job_schedule_type'] == 'Part-time')].copy()
df_pie= df_pie['job_schedule_type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle = 90, labels=['Full-time','Part-time'], explode=(0, 0.1), colors = cm.Reds([0.9, 0.7]))
df_pie.set_ylabel('')
df_pie.set_title('Job Schedule Full-time vs. Part-time')
```
