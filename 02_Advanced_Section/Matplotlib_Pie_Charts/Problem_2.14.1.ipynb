```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
from matplotlib import cm
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_da = df[df['job_title_short'] == 'Data Analyst']
df_da_pie = df_da['job_health_insurance'].value_counts().plot(kind='pie', explode=(0, 0.1),  autopct='%1.1f%%', startangle = 90, labels=['False','True'], colors = cm.Blues([0.9, 0.7]))
df_da_pie.set_ylabel('')
df_da_pie.set_title('Required Health Insurance for Data Analysist Jobs')
```
