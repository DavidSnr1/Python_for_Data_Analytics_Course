```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# plot 1
# In the first subplot, plot a bar chart of the top 3 job locations by the number of job postings.

df_1 = df.groupby(by='job_location')['job_posted_date'].count().sort_values(ascending=False).head(3)
```

```python
# plot 2
# In the second subplot, plot a horizontal bar chart of the top 3 job titles by the number of job postings.
df_2 = df.groupby('job_title_short')['job_posted_date'].count().head(3).sort_values(ascending=True)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(6, 6))
axs[0].bar(df_1.index,df_1.values)
axs[1].barh(df_2.index, df_2.values)
fig.tight_layout()
```
