```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
ger_jobs = df[df['job_country'] == "Germany"]
ger_jobs = ger_jobs[ger_jobs['salary_year_avg'].notna()]
ger_jobs.groupby('job_title_short')[['salary_year_avg']].max()
```
