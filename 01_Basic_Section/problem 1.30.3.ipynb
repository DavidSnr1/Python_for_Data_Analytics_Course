Matplotlib Labeling
Problem Statement:
Using the same line plot created in Matplotlib Plotting Question #3 showing the trend of average yearly salary over the first 6 months of the year:

Use the job_posted_date column to extract the month and calculate the average salary for each month.
Add labels for the x-axis (label as 'Month'), y-axis (label as 'Average Salary'), and a title (set as 'Average Salary by Job Month') for the plot.
Change the line color to 'red' and add a legend to the plot.
Hint:
Use the plt.plot() function to create the line plot and plt.xlabel(), plt.ylabel(), plt.title(), and plt.legend() to add labels, title, and legend.

```python
!pip install datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup (not needed for all problems but good to do anyways)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month']= df['job_posted_date'].dt.month
avg_salary_by_month = df.groupby('job_posted_month')['salary_year_avg'].mean().head(6)
avg_salary_by_month = avg_salary_by_month.sort_index(ascending = True)

plt.plot(avg_salary_by_month.index, avg_salary_by_month.values, label='Mean Salary', color = 'red')
plt.xlabel('Month')
plt.ylabel('Average Salary')
plt.title('Average Salary by Job Month')
plt.legend()
plt.show()
```
