Average Salary Horizontal Bar Chart (1.29.2) - Problem
Matplotlib Plotting
Problem Statement:
Create a horizontal bar chart showing the average yearly salary for different job titles. Use the job_title_short and salary_year_avg columns to calculate the average salary for each job title.

Hint:
Use the plt.barh() function to create the horizontal bar chart.

```python
!pip install datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup (not needed for all problems but good to do anyways)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
salary_for_job = df.groupby('job_title_short')['salary_year_avg'].mean()
salary_for_job = salary_for_job.sort_values(ascending = True)
plt.barh(salary_for_job.index, salary_for_job.values)
plt.show()
```
