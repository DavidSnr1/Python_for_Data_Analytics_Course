Problem Statement:
Group the DataFrame by job_country and calculate the total number of job postings for each country. Display the result.

Hint:
Use the groupby method to group the DataFrame by job_country and then use the size method to count the number of job postings for each group.

```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
total_postings_by_country = df.groupby('job_country')['job_country'].size().sort_values(ascending=False)
print(total_postings_by_country)
```
