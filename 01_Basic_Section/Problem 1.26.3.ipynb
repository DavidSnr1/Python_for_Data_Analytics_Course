Pandas Analysis
Problem Statement:
Group the DataFrame by job_title_short and calculate the average salary_hour_avg for each job title. Display the result.

Hint:
Use the groupby method to group the DataFrame by job_title_short and then use the mean method to calculate the average salary_hour_avg for each group.

```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
mean_byjobs = df.groupby('job_title_short')[['salary_hour_avg']].mean()
print(mean_byjobs)
```
