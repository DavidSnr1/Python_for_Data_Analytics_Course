Problem Statement:
Use a while loop to find the first job title that contains the word 'Engineer'. The list of job titles is ['Data Scientist', 'Data Analyst', 'Machine Learning Engineer', 'Software Engineer'].

Hint:
Use a while loop to iterate through the list and check if 'Engineer' is in each job title. Once you find the first matching job title, you'll want to stop the loop from continuing. Python has a special command called break that exits a loop immediately when you no longer need to continue searching.

```python
job_titles = ['Data Scientist', 'Data Analyst', 'Machine Learning Engineer', 'Software Engineer']

i = 0
while i < len(job_titles):
  if 'Engineer' in job_titles[i]:
    print(i ,job_titles[i])
    break
  else:
    i += 1
```
