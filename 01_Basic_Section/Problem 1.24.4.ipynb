Problem Statement:
Check for missing values in the job_skills column and display the rows where the values are not NA.

Hint:
Use the notna method to check for not NA values and boolean indexing to filter the DataFrame.

```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

not_na_values_df = df[df['job_skills'].notna()]
not_na_values_df
```
