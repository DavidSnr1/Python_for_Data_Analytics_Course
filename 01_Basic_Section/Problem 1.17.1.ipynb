Problem Statement:
Create a lambda function to calculate the average salary from a list of salaries. With salaries set as [95000, 120000, 105000, 90000, 130000].

Hint:
Use the sum and len functions within the lambda function to calculate the average.

```python
salaries = [95000, 120000, 105000, 90000, 130000]

avg = lambda avg: sum(salaries)/len(salaries)
avg_salary = avg(salaries)
print(avg_salary)
```
