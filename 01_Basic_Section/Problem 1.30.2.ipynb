Horizontal Bar Chart with Labels (1.30.2) - Problem
Matplotlib Labeling
Problem Statement:
Using the same bar chart created in Matplotlib Plotting Question #2 showing the average yearly salary for different job titles:

Use the 'job_title_short' and salary_year_avg columns to calculate the average salary for each job title.
Add labels for the x-axis (label as 'Average Salary') and a title (set as 'Average Salary by Job Title') for the plot.
Hint:
Use the plt.barh() function to create the horizontal bar plot and plt.xlabel() and plt.title() to add labels and titles.

```python
!pip install datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup (not needed for all problems but good to do anyways)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
yearly_salary_for_job = df.groupby('job_title_short')['salary_year_avg'].mean()
yearly_salary_for_job = yearly_salary_for_job.sort_values(ascending=True)

plt.barh(yearly_salary_for_job.index, yearly_salary_for_job.values,color = 'gray')
plt.xlabel("Average Salary")
plt.title("Average Salary by Job Title")
plt.show()
```
