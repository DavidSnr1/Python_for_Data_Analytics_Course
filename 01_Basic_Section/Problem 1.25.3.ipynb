Extract Year from Date (1.25.3) - Problem
Pandas Cleaning
Problem Statement:
Create a new column job_posted_year that extracts the year from the job_posted_date column and display the first 5 rows.

Hint:
Use the dt accessor to extract the year from the job_posted_date column.

```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_year'] = df['job_posted_date'].dt.year
df.head(5)
```
