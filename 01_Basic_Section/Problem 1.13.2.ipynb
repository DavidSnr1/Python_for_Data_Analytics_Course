Problem Statement:
Calculate the total number of job postings from a list using a while loop. The list of job postings is [5, 10, 15, 20, 25].

Hint:
Use a while loop to iterate through the list and sum the values. Use a variable index, set to 0, to iteratively go through the index of the list inside the while loop.

```python
job_postings = [5, 10, 15, 20, 25]
i = 0
while i < len(job_postings):
  allpostings += job_postings[i]
  i += 1
print(allpostings)
```
