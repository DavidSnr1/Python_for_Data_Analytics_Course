Problem Statement:
Use the os library to list all files in the current directory and print their names with the function os.listdir('.').

Hint:
Use the os.listdir function to get the list of files in the current directory, which is denoted by: '.'

```python
import os
file = os.listdir('.')
for files in file:
  print(files)
```
