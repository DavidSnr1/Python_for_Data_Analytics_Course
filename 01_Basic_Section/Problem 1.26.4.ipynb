Job with Lowest Hourly Salary (1.26.4) - Problem
Pandas Analysis
Problem Statement:
Find the job title with the lowest average hourly salary and display the job title and the salary.

Hint:
Use the idxmin method to find the index of the minimum value in the salary_hour_avg column and then use iloc to get the corresponding row.

```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
min_id = df['salary_hour_avg'].idxmin()
df.iloc[min_id][['job_title_short', 'salary_hour_avg']]
```
