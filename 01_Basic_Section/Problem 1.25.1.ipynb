Problem Statement:
Sort the DataFrame by salary_year_avg in descending order and display the first 10 rows.

Hint:
Use the sort_values method to sort the DataFrame by salary_year_avg in descending order.



```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df.sort_values(by='salary_year_avg', ascending=False, inplace=True)
df.head(10)
```
