Hint:
Use the filter function in combination with a lambda function that checks both conditions.

```python
job_postings = [
    {'title': 'Data Scientist', 'skills': ['Python', 'SQL'], 'remote': True},
    {'title': 'Data Analyst', 'skills': ['Excel', 'SQL'], 'remote': False},
    {'title': 'Machine Learning Engineer', 'skills': ['Python', 'TensorFlow'], 'remote': True},
    {'title': 'Software Developer', 'skills': ['Java', 'C++'], 'remote': True}
]
jobs = list(filter(lambda j: "Python" in j['skills'] and j['remote'] == True, job_postings ))
names = [thejob['title'] for thejob in jobs]
print(names)
```
