Problem Statement:
Find the union of two sets of job skills and print the result. The first set is {'Python', 'SQL', 'Tableau'} and the second set is {'R', 'SQL', 'Machine Learning'}.

Hint:
A union combines two sets and removes any duplicates, giving you all unique items from both sets. Python sets have a built-in method called union() to do this. The syntax is: set1.union(set2).

```python
skills_set1 = {'Python', 'SQL', 'Tableau'}
skills_set2 = {'R', 'SQL', 'Machine Learning'}
union = skills_set1.union(skills_set2)
print(union)
```
